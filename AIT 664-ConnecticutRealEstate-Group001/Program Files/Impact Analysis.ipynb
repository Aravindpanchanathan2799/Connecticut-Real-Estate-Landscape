```python
import pandas as pd

import numpy as np

pandas_df = pd.read_excel('/content/Complete.xlsx')

# Convert Pandas DataFrame to Spark DataFrame
df = spark.createDataFrame(pandas_df)

# Print schema
df.printSchema()
```

```python
def differenceOfItems(a, b):
        return round(a-b)
```

```python
def ratioOfItems(a, b):
    if a!=0 and b!=0:
        return round(b/a,2)
    else:
        return 0.00
```

```python
df_pandas = df.toPandas()
```

```python
def calculate_change(df, column_name):
    df[column_name + '_Change'] = (df.groupby('Town')[column_name].diff() /
                                   np.where(df.groupby('Town')[column_name].shift() != 0,
                                            df.groupby('Town')[column_name].shift(),
                                            np.nan)).round(2)
    df[column_name + '_Change'] = df[column_name + '_Change'].fillna(0)
    return df
```

```python
df.show()
```

```python
columns_to_change = ['Avg Number of Healthcare facilities', 'Avg Houses Sold (Condo)',  'Avg Condo Assessed Value', 'Avg Condo Sale Amount', 'Avg Condo Sale Ratio', 'Avg Houses Sold (Single)',  'Avg Single Family Assessed Value', 'Avg Single Family Sale Amount',  'Avg Single Family Sale Ratio', 'Avg Number of Schools', 'Avg Total Crimes',  'Avg Crimes per 100,000']
df_pandas = df_pandas.sort_values(['Town', 'List Year'])
for column in columns_to_change:
    df_pandas = calculate_change(df_pandas, column)
df_spark = spark.createDataFrame(df_pandas)
display(df_spark)
```

```python
df_spark.show()
```

```python
def calculate_impact_ratios(df, change_column, impact_column):
    impact_ratio_column = impact_column +"_"+ change_column +'_Impact_Ratio'
    df[impact_ratio_column] = (df[impact_column + '_Change'] /
                               np.where(df[change_column + '_Change'] != 0,
                                        df[change_column + '_Change'],
                                        np.nan)).round(2)
    df[impact_ratio_column] = df[impact_ratio_column].fillna(0)
    return df
df_spark = spark.createDataFrame(df_pandas)
display(df_spark)
```

```python
from sklearn.preprocessing import MinMaxScaler
changestobenormalised = ['Avg Number of Healthcare facilities_Change', 'Avg Houses Sold (Condo)_Change', 'Avg Condo Assessed Value_Change', 'Avg Condo Sale Amount_Change', 'Avg Condo Sale Ratio_Change', 'Avg Houses Sold (Single)_Change','Avg Single Family Assessed Value_Change','Avg Single Family Sale Amount_Change', 'Avg Single Family Sale Ratio_Change', 'Avg Number of Schools_Change', 'Avg Total Crimes_Change','Avg Crimes per 100,000_Change'  ]
change_columns = ['Avg Number of Healthcare facilities', 'Avg Number of Schools', 'Avg Total Crimes', 'Avg Crimes per 100,000']
impact_columns = ['Avg Houses Sold (Condo)', 'Avg Condo Assessed Value', 'Avg Condo Sale Amount',
                  'Avg Condo Sale Ratio', 'Avg Houses Sold (Single)', 'Avg Single Family Assessed Value',
                  'Avg Single Family Sale Amount', 'Avg Single Family Sale Ratio']
scaler = MinMaxScaler()
df_pandas[changestobenormalised] = scaler.fit_transform(df_pandas[changestobenormalised])
```

```python
df_pandas = df_pandas.sort_values(['Town', 'List Year'])
for change_column in change_columns:
    for impact_column in impact_columns:
        df_pandas = calculate_impact_ratios(df_pandas, change_column, impact_column)
        df_pandas = df_pandas.sort_values(['Town', 'List Year'])
impact_ratio_columns = [impact_column + '_' + change_column + '_Impact_Ratio' for change_column in change_columns for impact_column in impact_columns]
df_pandas[impact_ratio_columns] = scaler.fit_transform(df_pandas[impact_ratio_columns])
df_pandas
```
