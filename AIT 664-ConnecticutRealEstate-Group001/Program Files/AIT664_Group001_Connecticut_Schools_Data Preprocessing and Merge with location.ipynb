```python
import geopandas as gpd
import shapely
import pandas as pd
import numpy as np
from geopandas import GeoDataFrame


geojson_df = gpd.read_file("/content/Schools_Location.geojson")

geojson_df.count()
geojson_df.info()
geojson_df.head(20)
```

```python
df.show()
```

```python
geojson_df['geometry'] = geojson_df['geometry'].apply(lambda geom: geom.wkt)
```

```python
spark_geo_df = spark.createDataFrame(geojson_df)
```

```python
spark_geo_df.printSchema()
df.printSchema()
```

```python
df = df.withColumnRenamed("District Name", "District Name_existing")
df = df.withColumnRenamed("School Name", "School Name_existing")
df = df.withColumnRenamed("Organization Type", "Organization Type_existing")
df = df.withColumnRenamed("Organization Code", "Organization Code_existing")
df = df.withColumnRenamed("Address", "Address_existing")
df = df.withColumnRenamed("Town", "Town_existing")
df = df.withColumnRenamed("Zipcode", "Zipcode_existing")
df = df.withColumnRenamed("Phone", "Phone_existing")
df = df.withColumnRenamed("Student Open Date", "Student Open Date_existing")


```

```python
df.show()
```

```python
merged_df_spark = df.join(spark_geo_df, on="School ID", how="left")
```

```python
merged_df_spark.printSchema()
```

```python
# List of columns to drop
columns_to_drop = ['District Name', 'School Name', 'Organization Type', 'Organization Code',
                   'Address', 'Town', 'Zipcode', 'Phone', 'Student Open Date']

# Drop the specified columns
merged_df_spark = merged_df_spark.drop(*columns_to_drop)

# Show the resulting DataFrame
merged_df_spark.show()
```

```python
# List of columns to drop
columns_to_drop = ['result_num', 'osm_id', 'display_name', 'category', 'type', 'latlong']

# Drop the specified columns
merged_df_spark = merged_df_spark.drop(*columns_to_drop)

# Show the resulting DataFrame
merged_df_spark.show()
```

```python
from pyspark.sql.functions import col

# Remove '_existing' suffix from column names
merged_df_spark = merged_df_spark.select([col(c).alias(c.replace('_existing', '')) for c in merged_df_spark.columns])

# Show the DataFrame after renaming
merged_df_spark.show()
```

```python
for col_name in merged_df_spark.columns:
    # Filter and count null values and empty strings in the column
    null_count = merged_df_spark.filter(col(col_name).isNull() | (col(col_name) == "")).count()
    # Print the column name and corresponding null count
    print(f"Column: {col_name}\t Null Count: {null_count}")

# Count total rows in the DataFrame
total_count = merged_df_spark.count()
print(f"Total Count: {total_count}")
```

```python
# Filter out records where geometry is not null
merged_df_spark = merged_df_spark.filter(merged_df_spark.geometry.isNotNull())

# Show the DataFrame after filtering
merged_df_spark.show()
```

```python
for col_name in merged_df_spark.columns:
    # Filter and count null values and empty strings in the column
    null_count = merged_df_spark.filter(col(col_name).isNull() | (col(col_name) == "")).count()
    # Print the column name and corresponding null count
    print(f"Column: {col_name}\t Null Count: {null_count}")

# Count total rows in the DataFrame
total_count = merged_df_spark.count()
print(f"Total Count: {total_count}")
```

```python
# Repartition the DataFrame to have only one partition
merged_df_spark = merged_df_spark.repartition(1)

# Write the DataFrame to a new CSV file
merged_df_spark.write.csv("School_updated(1).csv", header=True)

# Download the CSV file to the system
from google.colab import files
files.download("School_updated(1).csv")
```
